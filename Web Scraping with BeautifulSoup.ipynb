```python
import requests
from bs4 import BeautifulSoup
import csv
import json

def scrape_website(url):
    # Send a GET request to the URL
    response = requests.get(url)
    # Parse HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find relevant information such as text, images, or links
    # For example, let's scrape all the headlines from a news website
    headlines = soup.find_all('h2', class_='headline')

    # Store the scraped data in a structured format (e.g., CSV or JSON)
    scraped_data = []
    for headline in headlines:
        # Extract text from the headline
        headline_text = headline.get_text()
        # Extract link from the headline
        headline_link = headline.find('a')['href']
        # Append the extracted data to the list
        scraped_data.append({'headline': headline_text, 'link': headline_link})

    return scraped_data

def save_to_csv(data, filename):
    # Save the data to a CSV file
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['headline', 'link']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for item in data:
            writer.writerow(item)

def save_to_json(data, filename):
    # Save the data to a JSON file
    with open(filename, 'w', encoding='utf-8') as jsonfile:
        json.dump(data, jsonfile, ensure_ascii=False, indent=4)

# Example usage:
url = 'https://products.basf.com/global/en/ci/n-vinyl-2-pyrrolidone.html'
scraped_data = scrape_website(url)
save_to_csv(scraped_data, 'scraped_data1.csv')
save_to_json(scraped_data, 'scraped_data1.json')
```
